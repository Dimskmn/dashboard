```python
# Mengubah tipe data menjadi categorical pada hour_df
hour_df['season'] = hour_df.season.astype('category')
hour_df['yr'] = hour_df.yr.astype('category')
hour_df['mnth'] = hour_df.mnth.astype('category')
hour_df['holiday'] = hour_df.holiday.astype('category')
hour_df['weekday'] = hour_df.weekday.astype('category')
hour_df['workingday'] = hour_df.workingday.astype('category')
hour_df['weathersit'] = hour_df.weathersit.astype('category')

hour_df.info()
```

**3. Mengubah value menjadi keterangan**

```python
# Mengubah value pada kolom mnth, season, weekday, weathersit, yr, workingday menjadi keterangan pada day_df berdasarkan readme.txt
day_df['mnth'] = day_df['mnth'].map({
    1: 'Jan', 2: 'Feb', 3: 'Mar', 4: 'Apr', 5: 'May', 6: 'Jun',
    7: 'Jul', 8: 'Aug', 9: 'Sep', 10: 'Oct', 11: 'Nov', 12: 'Dec'
})
day_df['season'] = day_df['season'].map({
    1: 'Spring', 2: 'Summer', 3: 'Fall', 4: 'Winter'
})
day_df['weekday'] = day_df['weekday'].map({
    0: 'Sun', 1: 'Mon', 2: 'Tue', 3: 'Wed', 4: 'Thu', 5: 'Fri', 6: 'Sat'
})
day_df['weathersit'] = day_df['weathersit'].map({
    1: 'Clear/Partly Cloudy',
    2: 'Misty/Cloudy',
    3: 'Light Snow/Rain',
    4: 'Severe Weather'
})

day_df['yr'] = day_df['yr'].map({
    0: '2011', 1: '2012'
})

day_df['workingday'] = day_df['workingday'].map({
    0: 'Holiday', 1: 'Workingday'
})

day_df.head()
```

```python
# Mengubah value pada kolom mnth, season, weekday, weathersit, yr, workingday menjadi keterangan pada hour_df berdasarkan readme.txt
hour_df['mnth'] = hour_df['mnth'].map({
    1: 'Jan', 2: 'Feb', 3: 'Mar', 4: 'Apr', 5: 'May', 6: 'Jun',
    7: 'Jul', 8: 'Aug', 9: 'Sep', 10: 'Oct', 11: 'Nov', 12: 'Dec'
})
hour_df['season'] = hour_df['season'].map({
    1: 'Spring', 2: 'Summer', 3: 'Fall', 4: 'Winter'
})
hour_df['weekday'] = hour_df['weekday'].map({
    0: 'Sun', 1: 'Mon', 2: 'Tue', 3: 'Wed', 4: 'Thu', 5: 'Fri', 6: 'Sat'
})
hour_df['weathersit'] = hour_df['weathersit'].map({
    1: 'Clear/Partly Cloudy',
    2: 'Misty/Cloudy',
    3: 'Light Snow/Rain',
    4: 'Severe Weather'
})

hour_df['yr'] = hour_df['yr'].map({
    0: '2011', 1: '2012'
})

hour_df['workingday'] = hour_df['workingday'].map({
    0: 'Holiday', 1: 'Workingday'
})

hour_df.head()
```

## Visualization & Explanatory Analysis

### Pertanyaan 1: Bagaimana tren jumlah penyewa sepeda dalam beberapa tahun terakhir?

```python
# Memvisualiasikan tren jumlah penyewa 1 tahun terakhir
day_df['month'] = pd.Categorical(day_df['month'], categories=
    ['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec'],
    ordered=True)

monthly_counts = day_df.groupby(by=["month","year"]).agg({
    "total_count": "sum"
}).reset_index()

sns.lineplot(
    data=monthly_counts,
    x="month",
    y="total_count",
    hue="year",
    palette="viridis_r",
    marker="o")

plt.title("Trend Jumlah Sewa Sepeda")
plt.xlabel(None)
plt.ylabel(None)
plt.legend(title="Tahun", loc="upper right")
plt.tight_layout()
plt.show()
```

**Insight:**
- Jumlah penyewa sepeda tertinggi terjadi pada bulan September 2012
- Jumlah penyewa sepeda terendah terjadi pada bulan Januari 2011
- Jumlah penyewa sepeda setiap bulan pada tahun 2012 lebih tinggi dibandingkan dengan jumlah penyewa pada bulan yang sama di tahun 2011

### Pertanyaan 2: Bagaimana jumlah penyewaan sepeda berdasarkan musim?

```python
sns.barplot(
    data=day_df,
    x="season",
    y="total_count",
    palette="viridis_r",
    errorbar=None)

plt.title("Jumlah Penyewaan Sepeda Berdasarkan Musim")
plt.ylabel(None)
plt.xlabel(None)
plt.tick_params(axis='x', labelsize=12)
plt.show()
```

- Sebaliknya jumlah penyewa sepeda paling sedikit terdapat pada musim semi (spring)

### Pertanyaan 3: Bagaimana jumlah penyewaan sepeda berdasarkan kondisi cuaca?

```python
weather_counts = day_df.groupby('weather')['total_count'].sum()
colors = sns.color_palette('viridis_r', n_colors=len(weather_counts))

plt.figure(figsize=(8, 8))
plt.pie(
    weather_counts,
    labels=weather_counts.index,
    autopct='%1.1f%%',
    colors=colors,
    startangle=90,
    wedgeprops={'edgecolor': 'white'}
)

plt.title('Jumlah Penyewa Sepeda berdasarkan Kondisi Cuaca')
plt.show()
```

### Pertanyaan 4: Bagaimana tren perbandingan penyewa dalam beberapa tahun terakhir?

```python
renters = day_df.groupby('year')[['casual', 'registered']].sum().reset_index()

# Ubah data dari format wide menjadi long untuk seaborn
renters_melted = renters.melt(id_vars='year', var_name='User Type', value_name='Count')

# Plot stacked bar chart menggunakan seaborn
plt.figure(figsize=(10, 6))
ax = sns.barplot(data=renters_melted, x='year', y='Count', hue='User Type', palette=['#3dbc74', '#238a8d'])

# Tambahan pengaturan
plt.xlabel(None)
plt.ylabel(None)
plt.title('Perbandingan Jumlah Penyewa Sepeda berdasarkan Pengguna (million)')
plt.legend(title='User Type')
plt.show()
```

**Insight:**
- Jumlah penyewa sepeda mengalami lonjakan pada tahun 2012
- Jumlah penyewa sepeda yang telah mendaftar meningkat signifikan pada tahun 2012 dibanding pada tahun 2011
- Jumlah penyewa sepeda yang belum mendaftar (casual) juga mengalami peningkatan pada tahun 2012

## Clustering Usage Category

Kita akan menggunakan metode binning atau quantile-based segmentation (qcut) untuk membagi hari-hari penyewaan ke dalam 3 kelompok, untuk bisa membandingkan dan melihat pola penyewaan pengguna.

*   Frequent Days: Hari dengan jumlah penyewaan tertinggi **(Top 33%)**
*   Occasional Days: Hari dengan penyewaan sedang **(Middle 33%)**
*   Casual Days: Hari dengan penyewaan terendah **(Bottom 33%)**



### Preparation

**Clustering user ke dalam 3 kelompok**

```python
# Buat kategori berdasarkan quantile (binning)
day_df['usage category'] = pd.qcut(day_df['total_count'], q=3, labels=['Casual Days', 'Occasional Days', 'Frequent Days'])

# Cek distribusi kategori
print(day_df['usage category'].value_counts())
```

### Visualization

**Distribusi Pengguna Berdasarkan Cluster Usage Category**

```python
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(30, 6))
colors = ["#D3D3D3", "#fcffa4", "#BA3B3D"]
sns.boxplot(x='usage category', y='registered', data=day_df, palette=colors, ax=ax[0])
ax[0].set_ylabel(None)
ax[0].set_xlabel(None)
ax[0].tick_params(axis='x', labelsize=30)
ax[0].tick_params(axis='y', labelsize=30)
ax[0].set_title("Registered User Distribution by Usage Category", fontsize=30)


sns.boxplot(x='usage category', y='casual', data=day_df, palette=colors, ax=ax[1])
ax[1].set_ylabel(None)
ax[1].set_xlabel(None)
ax[1].invert_xaxis()
ax[1].yaxis.tick_right()
ax[1].yaxis.set_label_position('right')
ax[1].tick_params(axis='x', labelsize=30)
ax[1].tick_params(axis='y', labelsize=30)
ax[1].set_title("Casual User Distribution by Usage Category", fontsize=30)
plt.show()
```

**Pola Penyewaan Pengguna berdasarkan Cluster Kategori**

```python
plt.figure(figsize=(10,5))
sns.countplot(x='weekday', hue='usage category', data=day_df, palette=["#fcffa4", "#ed6925", "#BA3B3D"])
plt.title("Customer Pattern by Usage Category in a week")
plt.tight_layout()
plt.legend(title="usage category", loc="lower right")
plt.xlabel(None)
plt.ylabel("Count")
plt.show()
```

## Conclusion

### Pertanyaan 1: Bagaimana tren jumlah penyewa sepeda dalam beberapa tahun terakhir?


Tren jumlah penyewa sepeda menunjukkan peningkatan dari tahun 2011 ke 2012. Setiap bulan di tahun 2012 memiliki jumlah penyewa yang lebih tinggi dibandingkan bulan yang sama di tahun 2011. Peningkatan paling signifikan terjadi mulai bulan April hingga Juni pada 2011 dan 2012, dan tren penyewaan mencapai puncaknya pada bulan September di tahun 2012. Setelah September, jumlah penyewa mulai menurun hingga akhir tahun. Secara keseluruhan, tren ini menunjukkan bahwa jumlah penyewa sepeda mengalami pertumbuhan dari tahun ke tahun.


### Pertanyaan 2: Bagaimana jumlah penyewaan sepeda berdasarkan musim?

Jumlah penyewaan sepeda bervariasi berdasarkan musim. Penyewaan sepeda tertinggi terjadi pada musim gugur (fall), diikuti oleh musim panas (summer). Sementara itu, jumlah penyewaan menurun pada musim dingin (winter), dan paling rendah terjadi pada musim semi (spring). Hal ini menunjukkan bahwa musim panas dan gugur adalah periode dengan aktivitas penyewaan sepeda yang lebih tinggi, kemungkinan karena cuaca yang lebih mendukung untuk bersepeda, sedangkan musim semi dan terutama musim dingin memiliki lebih sedikit penyewa, kemungkinan karena kondisi cuaca yang kurang mendukung.

### Pertanyaan 3: Bagaimana jumlah penyewaan sepeda berdasarkan kondisi cuaca?

Jumlah penyewaan tertinggi terjadi saat cuaca cerah atau berawan (Clear/Partly Cloudy). Sementara itu, pada kondisi hujan ringan atau bersalju (Light Snow/Rain), jumlah penyewa turun drastis hingga di bawah 2.000 orang. Hal ini menunjukkan bahwa cuaca berpengaruh signifikan terhadap tingkat penyewaan sepeda. Orang cenderung lebih banyak menyewa sepeda saat cuaca cerah atau mendung, sedangkan dalam kondisi hujan atau bersalju, minat untuk menyewa sepeda menurun secara signifikan.

### Pertanyaan 4: Bagaimana tren perbandingan penyewa dalam beberapa tahun terakhir?

Jumlah penyewa sepeda meningkat dari tahun 2011 ke 2012, baik untuk pengguna terdaftar (registered) maupun pengguna kasual (casual). Namun, peningkatan lebih signifikan terjadi pada penyewa terdaftar dibandingkan penyewa kasual. Hal ini menunjukkan tren bahwa semakin banyak pengguna yang memilih untuk mendaftar sebagai pelanggan tetap daripada hanya menyewa secara kasual. Secara keseluruhan, tren ini mencerminkan pertumbuhan jumlah pengguna sepeda, terutama di kalangan penyewa terdaftar.

### Pertanyaan 5: Pengguna mana yang lebih konsisten menyewa sepeda?

Pengguna terdaftar (registered users) lebih konsisten dalam menyewa sepeda dibandingkan pengguna kasual, karena memiliki distribusi penyewaan yang lebih stabil dibandingkan pengguna kasual. Mereka cenderung menyewa sepeda dalam jumlah yang lebih tinggi dan lebih konsisten, terutama pada kategori Frequent Days, dengan rentang yang lebih kecil dan median yang lebih tinggi. Sementara pengguna kasual (casual users) memiliki variabilitas yang lebih besar, terutama pada kategori Frequent Days yang menunjukkan penyebaran data yang lebih luas. Hal ini menunjukkan bahwa jumlah penyewaan pengguna kasual lebih fluktuatif dibandingkan pengguna terdaftar.

### Pertanyaan 6: Bagaimana pola pengguna melakukan penyewaan dalam satu minggu?

Pada akhir pekan (Sabtu & Minggu) jumlah penyewaan di dominasi oleh pengguna dengan kategori Casual Days dibandingkan hari kerja. Hal ini menunjukkan bahwa pengguna kategori Casual Days lebih cenderung menyewa sepeda untuk rekreasi atau aktivitas santai di akhir pekan. Pengguna dengan kategori Frequent Days lebih aktif pada hari kerja (Senin–Jumat) dibandingkan pengguna kasual. Ini mengindikasikan bahwa kelompok ini kemungkinan besar menggunakan sepeda sebagai sarana transportasi rutin.
